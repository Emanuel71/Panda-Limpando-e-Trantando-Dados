## Aula 01

```python
#Importando a base de dados 
import pandas as pd 

dados_churn = pd.read_json("arquivo.json")
dados_churn.head()
```

```python
#Normalizando todas colunas de uma vez
import json 

with open ('arquivo.json') as d:
    json_bruto = json.load(d)

dados_normalizado = pd.json_normalize(json_bruto)
dados_normalizado.head()
```

```python
dados_normalizado.info()
```

## Aula 02

```python
# Alterando o tipo da coluna 'conta.cobranca.Total' para 'float64'

dados_normalizado['conta.cobranca.Total'].astype(float)
```

```python
# Vamos verificar esse (' ') que está impossibilitando nossa alteração para 'float'
dados_normalizado[dados_normalizado['conta.cobranca.Total'] == ' '].head()
```

```python
# Selecionando colunas relevantes para correção 
dados_normalizado[dados_normalizado['conta.cobranca.Total'] == ' '][
    ['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']
]
```

```python
# Selecionando o idex de cada registro
idx = list(dados_normalizado[dados_normalizado['conta.cobranca.Total'] == ' '].index)
idx
```

```python
# Fazendo o calculo da cobrança total, pois sabemos o tempo do contrato '24'
dados_normalizado.loc[idx, 'conta.cobranca.Total'] = dados_normalizado.loc[idx, 'conta.cobranca.mensal'] * 24

dados_normalizado.loc[idx, 'cliente.tempo_servico']  = 24
```

```python
dados_normalizado.loc[idx][['cliente.tempo_servico', 'conta.contrato', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

```python
# Após fazer o tratamento, podemos fazer a converção do tipo
dados_normalizado['conta.cobranca.Total'] = dados_normalizado['conta.cobranca.Total'].astype(float)
```
