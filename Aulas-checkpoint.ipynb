## Aula 04

##### Identificando valores outliers

```python
#Verificando as estatistica do dataframe
df_sem_nulos.describe()
```

Sabendo que a coluna 'cliente.tempo_servico', é uma coluna de meses.
Esse valor de 90 anos (1080/12), é muito tempo de serviços contratos. Vamos analisar melhor, podemos encontras outliers

```python
#Importando a biblioteca Seaborn, para viasulizar melhor esses outliers 
import seaborn as sns

sns.boxplot(x=df_sem_nulos['cliente.tempo_servico'])
```

```python
#Definindo as medidas para localizar os 'outliers'

Q1 = df_sem_nulos['cliente.tempo_servico'].quantile(0.25)
Q3 = df_sem_nulos['cliente.tempo_servico'].quantile(0.75)
IIQ= Q3 - Q1
limite_inferior = Q1 - 1.5*IIQ
limite_superior = Q3 + 1.5*IIQ
```

```python
# Vamos criar um filtro e chamar apenas valores 'outliers'
outliers_index = (df_sem_nulos['cliente.tempo_servico'] < limite_inferior) | (df_sem_nulos['cliente.tempo_servico'] > limite_superior)
outliers_index
```

```python
df_sem_nulos.head()
```

```python
#Aplicamos filtro, apenas na coluna necessaria para analise 
df_sem_nulos[outliers_index]['cliente.tempo_servico']
```

Esses valores estão muito estranhos, porque não é possível assinar um serviço de telefonia por tanto tempo. Esses valores podem ser canditata a outliers.

#### Tratando valores outliers

```python
# Criando uma copia do dataframe
df_sem_out = df_sem_nulos.copy()
df_sem_out
```

Sabemos que o resultado da coluna 'cliente.tempo_servicos', é a divisão de duas colunas existente em nosso dataframe. Vamos realizar essa equação e tentar zera essas outliers

```python
df_sem_out.head(1)
```

```python
#Calculando os tempo de servicos
import numpy as np

df_sem_out.loc[outliers_index,'cliente.tempo_servico'] = np.ceil(
    (df_sem_out.loc[outliers_index,'conta.cobranca.Total']) / 
    (df_sem_out.loc[outliers_index,'conta.cobranca.mensal']))
```

```python
df_sem_out.head(1)
```

```python
sns.boxplot(x=df_sem_out['cliente.tempo_servico'])
```

```python
df_sem_out [outliers_index]['cliente.tempo_servico']
```

Podemos analiser acima, que ainda temos valores 'outliers'e outros que normalizaram.

```python
# Analisando as tres colunas.
df_sem_out[outliers_index][['cliente.tempo_servico', 'conta.cobranca.mensal', 'conta.cobranca.Total']]
```

Ao analisar os dados acima, podemos ver que os tres primeiros rigistro, continua 'outliers'. Pois, coluna 'conta.cobranca.Total' é outliers tambem, por isso o resultado continua com esse valor. Devemos fazer a remoção desses registroa


#### Removendo valores outliers

```python
#Filtra está comprometido, iremos fazer uma atualização para dataframe que estamos ultilizamos 
Q1 = df_sem_out['cliente.tempo_servico'].quantile(.25)
Q3 = df_sem_out['cliente.tempo_servico'].quantile(.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5*IQR
limite_superior = Q3 + 1.5*IQR

outliers_index = (df_sem_out['cliente.tempo_servico'] < limite_inferior) | (df_sem_out['cliente.tempo_servico'] > limite_superior)
outliers_index
```

```python
#Apenas registro outliers restantes
df_sem_out[outliers_index]
```

```python
# Vamos remover eles, fazendo uma negação desses registro acimas '~'
df_sem_out = df_sem_out[~outliers_index]
df_sem_out
```

```python
#Vamos redefinir os index, após a remoção dos outliers
df_sem_out.reset_index(drop=True, inplace=True)
```

```python
df_sem_out
```

#### Vamos remover e alterar informaçãoes, para performar melhor nosso modelo de analise

##### Removendo coluna

Iremos remover a coluna 'id_cliente', por ser dados unicos e não resistir nenhum padrão para analise. Se deixarmos essa coluna, nosso modelo perderá eficiencia

```python
df_sem_id = df_sem_out.drop('id_cliente', axis=1).copy()
df_sem_id
```

##### Categoria binaria

Ao analisar o data acima, percebemos que temos variáveis categóricas binárias. Isso quer dizer que temos colunas apenas com duas categoria, e normalmente deve deixar essas colunas como 0 e 1 para melhor analise, temos colunas que tem (não e sim) e (masc e fem).<br>
Vamos realizar a substituição desses valores para 0 e 1

```python
# Vamos mapear essas categoria
mapeamento = {'nao': 0, 'sim': 1,
              'masculino': 0, 'feminino': 1}
```

```python
# Vamos analisar quais colunas, devemos selecionar para realizar as substituição
for col in df_sem_id.columns:
    print(f"Coluna: {col}")
    print(df_sem_id[col].unique())
    print("-" * 30)
```

```python
colunas = ['conta.faturamente_eletronico', 'Churn', 'cliente.genero', 'cliente.parceiro', 
           'cliente.dependentes', 'telefone.servico_telefone']
```

```python
df_sem_id[colunas] = df_sem_id[colunas].replace(mapeamento)
```

```python
df_sem_id
```

##### Categoria não binaria

Para converter essas categoria não binaria, devemos ultilizar a função 'Dummies' do pandas. Essa função converte as categoria em colunas e valores delas em 0 e 1 ou True e False, dependendo da versão do Python usada. <br>
Com essa conversão, deixamos todas as colunas do dataframe com o tipo numerico, assim nosso modelo ML terá uma performace excelente

```python
#As colunas do tipo str, iram se tornar colunas numericas
df_sem_id.info()
```

```python
df_dummies = pd.get_dummies(df_sem_id).copy()
df_dummies
```
